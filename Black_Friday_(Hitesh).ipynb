```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(f'{pranavuikey_black_friday_sales_eda_path}/train.csv')
df
```

```python
df=df.dropna()
df
```

```python
df = df.drop_duplicates()
df
```

```python
# check for unique values in data
cat_cols = ['Gender', 'Age', 'City_Category', 'Stay_In_Current_City_Years', 'Marital_Status']
for col in cat_cols:
    print(f"{col}: {df[col].unique()}")
```

```python
df.info()
```

```python
df=df.reset_index(drop=True)
df
```

```python
df=df.drop(columns=['User_ID', 'Product_ID'])
df
```

```python
# perform EDA
sns.histplot(df['Purchase'], bins=50, kde=True)
plt.title('Purchase Amount Distribution')
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# check for outliers
sns.boxplot(y='Purchase',data=df)
plt.title("Outliers in Purchase Column")
plt.show()
```

```python
# Gender
df['Gender'] = df['Gender'].map({'F': 0, 'M': 1})

# Age
age_map = {'0-17': 0, '18-25': 1, '26-35': 2, '36-45': 3, '46-50': 4, '51-55': 5, '55+': 6}
df['Age'] = df['Age'].map(age_map)

# City_Category
df['City_Category'] = df['City_Category'].map({'A': 0, 'B': 1, 'C': 2})
```

```python
df
```

```python
# Age vs Purchased
sns.boxplot(x='Age',y='Purchase',data=df)
plt.xlabel('Ages')
plt.xlabel('Purchased')
plt.title('Purchase Distribution by Product Category 1')
plt.show()
```

```python
df.info()
```

```python
# Occupation vs Purchased
plt.figure(figsize=(16,6))
sns.barplot(x='Occupation',y='Purchase',data=df) #avg occupation display
plt.title('Purchase Distribution by Occupation')
plt.xticks(rotation=45)
plt.show()
```

```python
# Product_category_1 vs Purchased
plt.figure(figsize=(16,6))
sns.barplot(x='Product_Category_1', y='Purchase', data=df)#avg purchased
plt.title('Purchase Distribution by Product Category 1')
plt.xlabel('Product_Category_1')
plt.ylabel('Purchase')
plt.show()
```

```python
# Product_category_2 vs Purchased
plt.figure(figsize=(16,6))
sns.barplot(x='Product_Category_2',y='Purchase',data=df)
plt.title('Purchase Distribution by Product Category 2')
plt.xlabel('Product_Category_2')
plt.ylabel('Purchase')
plt.show()
```

```python
# Product_category_3 vs Purchased
plt.figure(figsize=(16,6))
sns.barplot(x='Product_Category_3',y='Purchase',data=df)
plt.title('Purchase Distribution by Product Category 3')
plt.xlabel('Product_Category_3')
plt.ylabel('Purchase')
plt.show()
```

```python
# City category pie chart
plt.figure(figsize=(8,8))
df['City_Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('City Category Distribution')
plt.ylabel('')
plt.show()
```

```python
# check for more possible plots
# 1. Gender vs Purchase
plt.figure(figsize=(16,6))
sns.barplot(x='Gender',y='Purchase',data=df,hue='Gender')
plt.title('Average Purchase by Gender')
plt.xlabel('Gender')
plt.ylabel('Purchase')
plt.show()
```

```python
plt.figure(figsize=(10,8))
numeric_df=df.select_dtypes(include='number')#only number col include
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
